```python
# Filtrar apenas por uma coluna

display(combustiveis_df['Revenda'])
```

```python
# DataFrame so com as colunas que eu quero

ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
# Exibe a 4a linha
display(ca_df.loc[3])

# Exibir da linha X at linha Y
display(ca_df.loc[9:19])
```

```python
# Criando dataFrame  contendo apenas as 4 colunas (revenda, municipio, produto, valor de venda), com combustivel sendo somente gasolina

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df[['Revenda', 'Produto', 'Valor de Venda']].max())
```

```python
# Preos de ETANOL em SP, ordenando do maior para o menor

eta_sp_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'SAO PAULO')]
eta_sp_df.sort_values(by='Valor de Venda', ascending=False, inplace=True)
display(eta_sp_df)
```

```python
# Mdia de preos de gasolina e gasolina aditivada do Bairro da Mooca
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), ['Valor de Venda']].mean())
```

```python
# Mostrar mdia de valor de venda por combustvel do Brasil todo

media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
# Adicionar uma coluna de valor booleano chamada 'Ativo' que por padro, vai ser Treue para TODAS as linhas

combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head(10))
```
