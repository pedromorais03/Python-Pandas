```python
# Importando o Pandas

import pandas as pd

combustiveis_df = pd.read_excel("/content/drive/MyDrive/ca202102_20230207120945.xlsx")
```

```python
display(combustiveis_df.head())
```

```python
# Inserção simples de dados
combustiveis_df['Ativo'] = True
display(combustiveis_df.head())
```

```python
# Criar uma coluna 'Obs' que tenha nela escrito 'Melhor cidade' quando 'Municipio' == 'SAO PAULO'
combustiveis_df['Obs'] = ['Melhor cidade' if municipio == 'SAO PAULO' else '' for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO', 'CAMPINAS', 'INDAIATUBA']), ['Municipio', 'Obs']])
```

```python
import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Valor de Venda - Status']])
```

```python
# Calcular postos de gasolina por habitantes temos na amostragerm de combustiveis nov/2021

ibge_df = pd.read_csv("/content/drive/MyDrive/ibge_num_habitantes_estimado.csv", sep=';')
ibge_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(ibge_df)
```

```python
# Fazer merge dos dois dataframes

colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(ibge_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
# Destruir coluna completamente vazia
merge_df.dropna(axis=1, inplace=True)
print(merge_df.info())
```

```python
colunas = ['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Unidade de Medida', 'Bandeira', 'Ativo', 'Obs']

merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
colunas2 = ['Valor de Venda - Status']

merge_df.drop(labels=colunas2, axis=1, inplace=True)
print(merge_df.info())
```

```python
# Remover linhas duplicadas

merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
# Agrupar e contar quantos postos tem na cidade

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={'Revenda':'Numero de Postos'}, inplace=True)
display(postos_por_municipio_df)
```
